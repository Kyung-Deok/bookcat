```python
import pandas as pd
# 파일 읽기
data = pd.read_csv("C:/Users/user/Desktop/Dashboard_data/원본/연평균 독서 권수.csv")
# 불필요한 열, 행 제거
df = data.iloc[:,:10]
df = df.iloc[:1]
# 불필요한 컬럼 제거
df = df.drop(['2004','2007'], axis=1)
# 데이터 값 바꾸기
df.loc[0,'Unnamed: 0']='독서량'
# 행, 열 바꾸기
df = df.transpose()
df.rename(columns=df.iloc[0],inplace=True)
# 불필요한 행 제거
df = df.drop(df.index[0])
# 인덱스 리셋
df.reset_index(inplace=True)
# 컬럼 이름 바꾸기
df.rename(columns = {'index':'년도'}, inplace = True)
df
```

```python
from sqlalchemy import create_engine
import pymysql
import pandas as pd
# mysql와 연결
db_connection_str = 'mysql+pymysql://root:1234@localhost/reading_database'
db_connection = create_engine(db_connection_str)
conn = db_connection.connect()
```

```python
#mysql에 저장
df.to_sql(name='avg_reading', con=db_connection, if_exists='replace', index=False)
```

```python
#추출한 데이터 엑셀 저장
df.to_excel("C:/Users/user/Desktop/Dashboard_data/D1.xlsx")
```
